```python
import matplotlib.pyplot as plt
import numpy as np

def kl_divergence(p, q):
    """
    Compute the Kullback-Leibler divergence between two probability mass vectors.

    Parameters:
    - p: numpy array representing the first probability mass function
    - q: numpy array representing the second probability mass function

    Returns:
    - kl_div: Kullback-Leibler divergence between p and q
    """
    # Ensure that the vectors have the same length
    assert len(p) == len(q), "Probability vectors must have the same length."

    # Avoid division by zero by adding a small constant to the vectors
    epsilon = 1e-20
    p = np.maximum(p, epsilon)
    q = np.maximum(q, epsilon)

    # Compute the KL divergence
    kl_div = np.sum(p * np.log2(p / q))

    return kl_div
def binary_entropy(p):
    """
    Compute binary entropy for a given probability.

    Parameters:
    - p: Probability (between 0 and 1)

    Returns:
    - entropy: Binary entropy for the given probability
    """
    if p == 0 or p == 1:
        return 0  # Entropy is 0 for perfectly certain events

    entropy = -p * np.log2(p) - (1 - p) * np.log2(1 - p)
    return entropy
def channel_rate(L,N,error_exp):
    error_hop = (1 - np.exp(- N * error_exp)) ** (L)
    if error_hop < 1/2:
        return 0
    else:
        #print("error hop: ", error_hop)
        Ent   = binary_entropy(error_hop)
        #print("entropy: ", Ent)
        if Ent > 1 or Ent < 0:
            print("error")
        #print("unnormalized rate: ", 1-Ent)
        rate = (1 - Ent)/N
        #print("rate: ", rate)
        return rate



epsilon = 0.05
Q       = np.array([[1-epsilon, epsilon], [epsilon, 1-epsilon]])
E_12    = -kl_divergence(Q[0,:], Q[1,:])
E_21    = -kl_divergence(Q[1,:], Q[0,:])
E_prime = np.minimum(E_12**2, E_21**2)
varrho  = epsilon
error_exp = E_prime / (2 * (np.log2(varrho)**2))


L_hist = np.arange(1, 1001)
N_hist = np.arange(1, 100)
rate_opt_hist = []
N_opt_hist    = []
for L in L_hist:
    rate_hist = []
    for N in N_hist:
        rate_NL = channel_rate(L,N,error_exp)
        rate_hist.append(rate_NL)

    rate_hist    = np.array(rate_hist)
    index_of_max = np.argmax(rate_hist)
    N_opt        = N_hist[index_of_max]
    rate_opt     = np.max(rate_hist)

    #print("Hop: ", L, "Nopt: ", N_opt, "rate_opt: ", rate_opt)
    rate_opt_hist.append(rate_opt)
    N_opt_hist.append(N_opt)

np.save("data/rate_opt_hist_epsilon_005.npy", np.array(rate_opt_hist))
np.save("data/N_opt_hist_epsilon_005.npy",    np.array(N_opt_hist))


```
